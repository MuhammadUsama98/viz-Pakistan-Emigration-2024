```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV
file_path = './Pakistan_Emigration_Remittance_Data_2023_2024.csv'
df = pd.read_csv(file_path)

# Visualization 1: Total Emigrants in 2023 and 2024
plt.figure(figsize=(6, 4))
plt.bar(["2023", "2024"], [862625, 727381], color="skyblue", edgecolor="black")
plt.title("Total Emigrants in 2023 vs 2024", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Emigrants", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Percentage Change in Emigrants
plt.figure(figsize=(8, 6))
plt.bar(["Emigrants Change"], [(-15)], color="lightcoral", edgecolor="black")
plt.title("Percentage Change in Emigrants (2023 to 2024)", fontsize=16)
plt.ylabel("Percentage Change", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Emigration by Profession (2024)
professions = ["Skilled Workers", "Unskilled Workers", "Highly Skilled Workers", "Other Professions"]
percentages = [37, 46, 5, 12]

plt.figure(figsize=(8, 4))
plt.bar(professions, percentages, color="lightgreen", edgecolor="black")
plt.title("Emigration by Profession (2024)", fontsize=16)
plt.xlabel("Profession", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Emigrants by Country (2024)
countries = ["Saudi Arabia", "Oman", "UAE", "Qatar", "Bahrain", "UK", "Malaysia", "USA", "Other Countries"]
emigrants_by_country = [452562, 81587, 64130, 40818, 25198, 13695, 13345, 1077, 34969]

plt.figure(figsize=(8, 6))
plt.bar(countries, emigrants_by_country, color="skyblue", edgecolor="black")
plt.title("Emigrants by Country (2024)", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Emigrants", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization 5: Remittances by Country (December 2024)
remittances_by_country = [770.6, 631.5, 456.9, 360.3, 284.3, 575.7]
countries_for_remittances = ["Saudi Arabia", "UAE", "UK", "EU", "USA", "Other Countries"]

plt.figure(figsize=(10, 6))
plt.bar(countries_for_remittances, remittances_by_country, color="lightgreen", edgecolor="black")
plt.title("Remittances by Country (December 2024)", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Remittances (Million USD)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization 6: Total Remittances in 2023 vs 2024
total_remittances = [26.359, 34.634]

plt.figure(figsize=(6, 4))
plt.bar(["2023", "2024"], total_remittances, color="lightblue", edgecolor="black")
plt.title("Total Remittances (2023 vs 2024)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Remittances (Billion USD)", fontsize=12)
plt.tight_layout()
plt.show()
```
